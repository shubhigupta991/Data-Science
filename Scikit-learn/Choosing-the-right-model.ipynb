### Scikit-learn uses estimator as another term for machine learning model or algorithm

* Classification :- predicting whether a sample is one thing or other
* Regression :- predicting a number

For finding the model try :- https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html

### Choosing the right estimator for Regression problem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
boston
```

```python
boston_df = pd.DataFrame(boston['data'],columns = boston['feature_names'])
boston_df['target'] = pd.Series(boston['target'])
```

```python
len(boston_df)
```

```python
boston_df.head()
```

```python
#Trying Ridge Regression model
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

#Setting up random seed
np.random.seed(0)

#Creating data
X = boston_df.drop('target', axis = 1)
y = boston_df['target']

#Splitting data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

#Instantiating model
model_1 = Ridge()
model_1.fit(X_train, y_train)

#Check the score of Ridge model on test data
model_1.score(X_test, y_test)
```

How to improve this?

Check :-https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html

```python
#Trying the same data with RandomForest
from sklearn.ensemble import RandomForestRegressor

#Setting random seed
np.random.seed(1)

#Create data
X = boston_df.drop('target', axis = 1)
y = boston_df['target']

#Splitting data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)

#Instantiating model
model_2 = RandomForestRegressor()
model_2.fit(X_train,y_train)

#Check the score of random forest model on test data
model_2.score(X_train,y_train)
```

### Choosing the right model for classification problem.

```python
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease.head()
```

```python
len(heart_disease)
```

```python
#Trying LinearSVC model
from sklearn.svm import LinearSVC

#Create data
X = heart_disease.drop('target',axis=1)
y = heart_disease['target']

#Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)

#Instantiating model
clf_1 = LinearSVC()
clf_1.fit(X_train,y_train)

#Check the score on test data
clf_1.score(X_test,y_test)
```

```python
#Trying same data with RandomForest model
from sklearn.ensemble import RandomForestClassifier

#Create data
X = heart_disease.drop('target',axis=1)
y = heart_disease['target']

#Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)

#Instantiating model
clf_2 = RandomForestClassifier()
clf_2.fit(X_train,y_train)

#Check the score on test data
clf_2.score(X_test,y_test)
```
