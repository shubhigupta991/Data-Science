```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
boston
```

```python
boston_df = pd.DataFrame(boston['data'], columns = boston['feature_names'])
boston_df['target'] = pd.Series(boston['target'])

boston_df.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X = boston_df.drop('target', axis = 1)
y = boston_df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

model = RandomForestRegressor()

model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

## Regression Model Evaluation metrices

* r2_score
* Mean absolute error(MAE)
* Mean squared error(MSE)

**r2_score**

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_test)
```

```python
y_preds = model.predict(X_test)
```

```python
r2_score(y_test,y_preds)
```

**Mean Absolute Error(MAE)**

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, y_preds)
```

What is mean absolute error?

```python
df = pd.DataFrame({'True values': y_test, 'Predicted values' : y_preds})
df.head()
```

```python
df['Difference'] = df['Predicted values'] - df['True values'] 
df
```

**Mean Squared Error(MSE)**

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_preds)
```

```python
df['Squared Difference'] = np.square(df['Difference'])
df
```

```python
np.mean(df['Squared Difference'])
```
