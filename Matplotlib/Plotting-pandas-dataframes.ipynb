# Which one should we use? (pyplot VS matplotlib OO method)

* When plotting something quickly, use the pyplot method.
* When plotting something advanced, use the OO method.

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind = 'scatter',
            x='age',y = 'chol',
            c = 'target');      
```

```python
#pyplot mixed with OO method
fig,ax = plt.subplots()
over_50.plot(kind = 'scatter' ,figsize = (10,6),
                     x = 'age', y = 'chol',
                     c = 'target' , ax = ax);
```

```python
##### OO method from scratch
fig,ax = plt.subplots(figsize = (10,6))

#Plotting data
scatter = ax.scatter(
                 x = over_50['age'], y = over_50['chol'],
                 c = over_50['target'])
#Customising plot
ax.set(title = 'Heart-Disease And Cholestrol Levels',
      xlabel = 'Age', ylabel = 'Cholestrol')
#Add a legend
ax.legend(*scatter.legend_elements(),title = 'Target')

#Add a horizontal line
ax.axhline(over_50['chol'].mean() ,linestyle = '--');
```

```python
fig, (ax0,ax1) = plt.subplots(nrows = 2, ncols = 1,
                             figsize = (10,10), sharex = True)
#Add data to ax0
scatter = ax0.scatter(x=over_50['age'],y=over_50['chol'],
                      c=over_50['target'])

#Customising ax0
ax0.set(title = 'Heart-Diseases And Cholestrol level',
        ylabel = 'Cholestrol')

#Add a legend
ax0.legend(*scatter.legend_elements(),title = 'Target')

#Add a horizontal line
ax0.axhline(heart_disease['chol'].mean(),linestyle = '--')

#Add data to ax1
scatter = ax1.scatter(x=heart_disease['age'],y=heart_disease['thalach'],
                     c=heart_disease['target'])

#Customising ax1
ax1.set(title = 'Heart-Diseases And Max Heart rate ',
       xlabel = 'Age',ylabel = 'Max Heart rate')

#Add a legend
ax1.legend(*scatter.legend_elements(),title = 'Target')

#Add a horizontal line
ax1.axhline(heart_disease['thalach'].mean(),linestyle = '--')

#Add title to figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```
